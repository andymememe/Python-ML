```python
import numpy as np
import matplotlib.pyplot as plt
from classifier.Perceptron import Perceptron
from shared import Plot, Data
```

```python
# Get Iris data
print('Download Iris data...')
iris = Data.get_iris_data()
print(iris.tail())
print('Finished!')
```

```python
# Get training Data
y = iris.iloc[0:100, 4].values  # Get species
y = np.where(y == 'Iris-setosa', -1, 1)  # Iris-setosa : -1, else : 1
X = iris.iloc[0:100, [0, 2]].values  # Get sepal length & peral length
```

```python
# Plot data
print('\nPlot data...')
plt.scatter(X[:50, 0], X[:50, 1], color='red', marker='o',
            label='setosa')
plt.scatter(X[50:100, 0], X[50:100, 1], color='blue', marker='x',
            label='versicolor')
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.legend(loc='upper left')
plt.show()
```

```python
# Perceptron
ppn = Perceptron(eta=0.1, n_iter=10)
ppn.fit(X, y)
```

```python
# Plot errors
print('\nPlot errors...')
plt.plot(range(1, len(ppn.errors_) + 1), ppn.errors_, marker='o')
plt.xlabel('Epochs')
plt.ylabel('# of misclassifications')
plt.show()
```

```python
# Plot result
print('\nPlot result...')
Plot.plot_decision_regions(X, y, classifier=ppn)
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.legend(loc='upper left')
plt.show()
```
